```python
# 업종 변경 업체 존재함 -> 아래에서 확인 
# loca_unique와 loca_type_unique에서 겹치지 않는 애들 
change_store = pd.concat([loca_unique, loca_type_unique])
change_store = change_store.reset_index(drop = True)

change_store_g = change_store.groupby(list(change_store.columns))

idx = [x[0] for x in change_store_g.groups.values() if len(x) == 1]
change_store = change_store.reindex(idx)
```

#### 좌표계 변환 | ITRF2000 -> KaKao 기준 WGS84

```python
from pyproj import Proj, transform
```

```python
# ITRF2000
proj_ITRF2000=Proj(init='epsg:5178')
#WGS84
proj_WGS84=Proj(init='epsg:4326')

#좌표 변환
x, y = list(data_concat['POINT_X']), list(data_concat['POINT_Y'])
X, Y = transform(proj_ITRF2000, proj_WGS84, x, y)

data_concat['WGS_X'] = X
data_concat['WGS_Y'] = Y
data_concat
```
